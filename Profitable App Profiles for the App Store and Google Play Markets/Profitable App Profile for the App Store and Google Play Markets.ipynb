Next, we will remove the dublicates. If we look at the entries for Instagram we can see that the reviews column is different in each row. We will only keep the highest one, because it's the most recent one. We will keep the row with the highest number of reviews.

```python
print('Expected length:', len(google_data[1:])-len(dublicate_apps))
```

To do that we have to create a dictionary each key is a unique app name and the corresponding dictionary value is the highest number of reviews of that app.

```python
reviews_max = {}
for app in google_data[1:]:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

```python
print(len(reviews_max))
```

We will use the above dictionary to remove dublicate rows

```python
android_clean = []
already_added = []

for row in google_data[1:]:
    name = row[0]
    n_reviews = float(row[3])
    
    if (n_reviews == reviews_max[name]) and (name not in already_added):
        android_clean.append(row)
        already_added.append(name)
```

Next, we will remove the non-English apps from our dataset. There is a function below that detects non-English apps.

```python
def is_english(string):
    non_ascii = 0
    
    for i in string:
        if ord(i) > 127:
            non_ascii +=1
    if non_ascii >3:
        return False
    else:
        return True
```

In the next steps we want to find most common genres of apps in both markets. If an app is in both markets, then the app is probably profitable. Therefore we want to analyze most common genres in both markets.

```python
print(apps_data[0][11])
print(google_data[0][9])
print(google_data[0][1])
```

```python
def freq_table(dataset, index):
    freq = {}
    total = 0
    
    for row in dataset:
        total +=1
        value = row[index]
        if value in freq:
            freq[value] += 1
        else:
            freq[value] = 1
    
    percentages = {}
    for i in freq:
        perc = (freq[i] / total)*100
        percentages[i] = perc
    return percentages
```

Below, there is a function provided by Dataquest:

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
print(display_table(apple_free, 11))
```

The proportion of Games is more than half of all free English apps. The free English apps are mostly for entertainment and hobbies. I would recommend a basic, free but addicting game app to build. We analyzed the data of Apple Store, next we will analyze Google Play Store data from the number of installs.

```python
google_categories = freq_table(google_free,1)
```

```python
for category in google_categories:
    total = 0         #sum of installs specific to each genre
    len_category = 0  #number of apps specific to each genre
    for app in google_free:
        category_app = app[1]
        if category_app == category:
            installs = app[5]
            installs = installs.replace('+', '')
            installs = installs.replace(',','')
            installs = float(installs)
            total += installs
            len_category += 1
    avg = total / len_category
    print(category, ":", installs)
```
