Initial Exploring and Cleaning
---

```python
autos.describe()
```

As we can see above, the nr_of_pictures columns doesn't have any valuable data. Also some values shouldn't have decimal places like months, or registration years. Price and odometer datas should also be in here since they are numerical. But first, they need cleaning from text and then we will convert them into numerical data.

First we will remove $ and comma from the price column.

```python
autos["price"] = (autos["price"]
                  .str.replace("$","")
                  .str.replace(",","")
                  .astype(int))
autos["price"].unique()
```

Next, we will remove km and comma from odometer column values and also change the column name to odometer_km.

```python
autos["odometer"] = (autos["odometer"]
                             .str.replace("km","")
                             .str.replace(",","")
                             .astype(int)
                             )
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
autos["odometer_km"].head()
```

Dealing with Incorrect Registration Year Data
---

```python
autos["registration_year"].describe()
```

The minimum registration year is 1000 which is an odd value, many years before the cars were invented, and the maximum year is 9999 which is many years ahead from today.

We need to clear some data that fall outside of the years 1900-2016.

```python
autos["registration_year"].value_counts(normalize=True).sort_index()
```

```python
(autos["registration_year"].between(1900,2016).sum())/autos.shape[0]
```

The percentage of the cars that are inside the range of 1900 and 2016 is 96.1%.

```python
(~autos["registration_year"].between(1900,2016)).sum() / autos.shape[0]
```

The percentage of the car that are outside of our range is 3.9%. Therefore we will remove this data.

```python
autos = autos[autos["registration_year"].between(1900,2016)]
```

```python
autos["registration_year"].value_counts(normalize=True).head(10)
```

```python
(autos["registration_year"]
 .value_counts(normalize=True)
 .sort_values(ascending = False)
 .head(20))
```

We sorted the registration years by their percentages. We can see that the most of the cars were registered in the past 20 years.

Exploring Price by Brand
---

Now let's explore the brand column.

```python
autos["brand"].unique()
```

```python
(autos["brand"]
 .value_counts(normalize=True)
 .head(10)
 .sort_values(ascending=False))
```

```python
brand_names = autos["brand"].value_counts().head(10).index
```

```python
brand_names
```

We will continue our analysis with these top 10 brands since they are the most common brands on our dataset.

```python
brand_prices = {}

for i in brand_names:
    name = autos[autos["brand"] == i]
    mean_price = name["price"].mean()
    brand_prices[i] = int(mean_price)
```

```python
brand_prices
```

It looks like Audi, BMW, Mercedes-Benz have the highest prices, Volkswagen is in between, and others have lower prices.

Next, we will calculate the mean milage for these brands.

```python
brand_milage = {}

for i in brand_names:
    name = autos[autos["brand"] == i]
    mean_milage = name["odometer_km"].mean()
    brand_milage[i] = int(mean_milage)
    
brand_milage
```

It looks like on average, there are not much milage differences between the car brands.

Storing Aggregate Data in a DataFrame
---

Next, we will convert the two dictionaries into pandas series.

```python
brand_p = pd.Series(brand_prices)
```

```python
brand_m = pd.Series(brand_milage)
```

```python
df_price_milage = pd.DataFrame(brand_p, columns = ["mean_price"])
```

```python
df_price_milage["mean_milage"] = brand_m
```

```python
df_price_milage
```

As we can see on our dataframe, the brand affects the average price, but the average milage, doesn't affect the prices of the cars.
For example, Opel has 129310 milage on average and Audi has 129157 milage on average. The values are almost the same, however there is a huge price gap between these two brands.
