Visualizing Earnings Based On College Majors
===

This project visualizes the data of students who graduated college between 2010 and 2020 by their majors and outcomes. We used scatter plots, histograms and bar plots to visualize the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
```

Further Exploration
---

A grouped bar plot to compare the number of men with the number of women in each category of majors.

```python
recent_grads.groupby('Major_category')['Men', 'Women'].agg('sum').plot.bar();
```

Box plots to explore distributions of median salaries and unemployment rate.

```python
recent_grads['Unemployment_rate'].plot.box()
```

```python
recent_grads['Median'].plot.box()
```

```python
highest_75 = recent_grads.groupby(["Major"])["P75th"].agg("sum").sort_values(ascending=False).head()
highest_75
```

```python
plt.figure(figsize=(8,5))
highest_75.plot.bar()
plt.title("Top 5 Majors with the Highest 75th Percentile of Earnings")
```

```python
highest_fulltime = recent_grads.groupby(["Major"])["Full_time"].agg("sum").sort_values(ascending=False).head()
highest_fulltime
```

```python
plt.figure(figsize=(8,5))
highest_fulltime.plot.bar()
plt.title("Top 5 Majors with the Highest Fulltime Employees")
```

```python
lowest_median = recent_grads.groupby(["Major"])["Median"].agg("sum").sort_values().head()
lowest_median
```

```python
plt.figure(figsize=(8,5))
lowest_median.plot.bar()
plt.title("Top 5 Majors with the Lowest Wages")
```
